```python
import torch
import matplotlib.pyplot as plt
import numpy as np
from torch import nn
```

```python
# Generating data (Sine curve)
T = 1000
Time = torch.arange(1,T+1)
x = torch.sin(0.01*Time)+torch.normal(0,0.2,(T,))
plt.plot(Time,x)
plt.xlabel('Time')
plt.ylabel('X')
plt.title('Sequence Data')
plt.show()
```

```python
tau = 4
n_train = 600
```

```python
np.shape(x[:tau])
```

```python
Features = np.zeros((T-tau,tau))
Label = np.zeros((T-tau,1))
for i in range(np.shape(Features)[0]):
    
    Features[i,:] = x[i:i+tau]
    Label[i,0] = x[i+tau]
```

```python
def init_weights(m):
    
    if typr(m) == nn.Linear:
        nn.init.xavier_uniform_(m.weight)
        
def net():
    
    n_net = nn.Sequential(nn.Linear(4, 10), nn.ReLU(), nn.Linear(10, 1))
    
    n_net.apply(init_weights)
    
loss = nn.MSELoss()
```

```python
network = net()
trainer = torch.optim.Adam(network.parameters())
```

```python
train_feature, train_label = Features[:n_train,:],Label[:n_train]
```
